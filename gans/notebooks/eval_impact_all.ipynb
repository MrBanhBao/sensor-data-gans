```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
standing_df = pd.read_hdf('./data/standing.h5')
walking_df = pd.read_hdf('./data/walking.h5')
jogging_df = pd.read_hdf('./data/jogging.h5')
```

```python
df = pd.concat([standing_df, walking_df, jogging_df])
```

```python
plot = sns.lineplot(x='percentage', y='test_f1', hue='action', data=df)
```

```python
plot.figure.savefig("test_impact.pdf")
```
